```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

#Extract the edges from an image with kernel (Sobel and Prewitt)

```python
# Function to apply convolution using a kernel
def apply_convolution(image, kernel):
    return cv2.filter2D(image, -1, kernel)

# Upload your image (replace 'path_to_image' with your image path)
image = cv2.imread('/gdrive/MyDrive/image_processing/The-famous-Lena-image-often-used-as-an-example-in-image-processing.png', cv2.IMREAD_GRAYSCALE)

# Define different kernels for edge detection
sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])  # Sobel kernel for X-axis edges
sobel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])   # Sobel kernel for Y-axis edges
prewitt_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]]) # Prewitt kernel for X-axis edges
prewitt_y = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]]) # Prewitt kernel for Y-axis edges

# Apply convolution using the defined kernels
edges_sobel_x = apply_convolution(image, sobel_x)
edges_sobel_y = apply_convolution(image, sobel_y)
edges_prewitt_x = apply_convolution(image, prewitt_x)
edges_prewitt_y = apply_convolution(image, prewitt_y)

# Display the results
plt.figure(figsize=(12, 12))
plt.subplot(221), plt.imshow(edges_sobel_x, cmap='gray'), plt.title('Sobel X')
plt.subplot(222), plt.imshow(edges_sobel_y, cmap='gray'), plt.title('Sobel Y')
plt.subplot(223), plt.imshow(edges_prewitt_x, cmap='gray'), plt.title('Prewitt X')
plt.subplot(224), plt.imshow(edges_prewitt_y, cmap='gray'), plt.title('Prewitt Y')
plt.show()
```


#Adding 3 different type of blurring

```python
# Function to apply convolution for blurring
def apply_blur(image, kernel):
    return cv2.filter2D(image, -1, kernel)

# Load a sample image
image = cv2.imread('/gdrive/MyDrive/image_processing/The-famous-Lena-image-often-used-as-an-example-in-image-processing.png', cv2.IMREAD_GRAYSCALE)

# Define different blurring kernels
# 1. Large averaging kernel (15x15)
blur_kernel_large = np.ones((15, 15), np.float32) / 225
# 2. Gaussian blur kernel (15x15)
blur_kernel_gaussian = cv2.getGaussianKernel(15, 5)
blur_kernel_gaussian = np.dot(blur_kernel_gaussian, blur_kernel_gaussian.T)
# 3. Median blur

# Apply blurring using the defined kernels
blurred_image_large = apply_blur(image, blur_kernel_large)
blurred_image_gaussian = apply_blur(image, blur_kernel_gaussian)
blurred_image_median = cv2.medianBlur(image, 15)

# Display the original and blurred images
plt.figure(figsize=(15, 10))
plt.subplot(221), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(222), plt.imshow(blurred_image_large, cmap='gray'), plt.title('Large Averaging Blur')
plt.subplot(223), plt.imshow(blurred_image_gaussian, cmap='gray'), plt.title('Gaussian Blur')
plt.subplot(224), plt.imshow(blurred_image_median, cmap='gray'), plt.title('Median Blur')
plt.show()
```

#Ploting the difference pictures between the original image to the blured images

```python
# Function to apply convolution for blurring
def apply_blur(image, kernel):
    return cv2.filter2D(image, -1, kernel)

# Function to calculate the difference between two images
def calculate_difference(image1, image2):
    # Taking the absolute difference between the original and blurred image
    return cv2.absdiff(image1, image2)

# Load a sample image
image = cv2.imread('/gdrive/MyDrive/image_processing/The-famous-Lena-image-often-used-as-an-example-in-image-processing.png', cv2.IMREAD_GRAYSCALE)

# Define different blurring kernels
# 1. Large averaging kernel (15x15)
blur_kernel_large = np.ones((15, 15), np.float32) / 225
# 2. Gaussian blur kernel (15x15)
blur_kernel_gaussian = cv2.getGaussianKernel(15, 5)
blur_kernel_gaussian = np.dot(blur_kernel_gaussian, blur_kernel_gaussian.T)
# 3. Median blur (applied separately as it's not a convolution)
# For median blur, the kernel size must be a positive odd integer

# Apply blurring using the defined kernels
blurred_image_large = apply_blur(image, blur_kernel_large)
blurred_image_gaussian = apply_blur(image, blur_kernel_gaussian)
blurred_image_median = cv2.medianBlur(image, 15)

# Calculate differences
difference_large = calculate_difference(image, blurred_image_large)
difference_gaussian = calculate_difference(image, blurred_image_gaussian)
difference_median = calculate_difference(image, blurred_image_median)

# Display the differences
plt.figure(figsize=(15, 10))
plt.subplot(221), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(222), plt.imshow(difference_large, cmap='gray'), plt.title('Difference with Large Averaging Blur')
plt.subplot(223), plt.imshow(difference_gaussian, cmap='gray'), plt.title('Difference with Gaussian Blur')
plt.subplot(224), plt.imshow(difference_median, cmap='gray'), plt.title('Difference with Median Blur')
plt.show()


plt.show()
```

#plot the histogram equalization of the image

```python
# Function to load an image from a file
def load_image(file_path):
    return cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)

# Function to perform histogram equalization
def equalize_histogram(image):
    return cv2.equalizeHist(image)

# Load the image
image_path = '/gdrive/MyDrive/image_processing/The-famous-Lena-image-often-used-as-an-example-in-image-processing.png'  # Replace with your image path
image = load_image(image_path)

# Perform histogram equalization
equalized_image = equalize_histogram(image)

# Calculate histograms
hist_original = cv2.calcHist([image], [0], None, [256], [0, 256])
hist_equalized = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])

# Plot the images and histograms
plt.figure(figsize=(12, 6))

# Original image and histogram
plt.subplot(221), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(222), plt.plot(hist_original), plt.title('Histogram of Original Image')

# Equalized image and histogram
plt.subplot(223), plt.imshow(equalized_image, cmap='gray'), plt.title('Histogram Equalized Image')
plt.subplot(224), plt.plot(hist_equalized), plt.title('Histogram of Equalized Image')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to load an image
def load_image(file_path):
    return cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)

# Function to apply a high-pass filter in the frequency domain
def high_pass_filter(image):
    # Perform the Fourier transform
    f = np.fft.fft2(image)
    fshift = np.fft.fftshift(f)

    # Create a magnitude spectrum
    magnitude_spectrum = 20 * np.log(np.abs(fshift))

    # Get the image dimensions
    rows, cols = image.shape
    crow, ccol = rows // 2 , cols // 2

    # Create a mask with a high pass filter (1 at the corners, 0 at the center)
    mask = np.ones((rows, cols), np.uint8)
    mask[crow-30:crow+30, ccol-30:ccol+30] = 0

    # Apply the mask and inverse DFT
    fshift_masked = fshift * mask
    f_ishift = np.fft.ifftshift(fshift_masked)
    img_back = np.fft.ifft2(f_ishift)
    img_back = np.abs(img_back)

    return img_back, magnitude_spectrum, fshift_masked

# Load your image (replace with your image path)
image_path = '/gdrive/MyDrive/image_processing/The-famous-Lena-image-often-used-as-an-example-in-image-processing.png'  # Replace with the path to your image
image = load_image(image_path)

# Apply high pass filter
high_passed_image, magnitude_spectrum, filtered_spectrum = high_pass_filter(image)

# Display the original image, magnitude spectrum, filtered spectrum and high-pass filtered image
plt.figure(figsize=(16, 8))

plt.subplot(141), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(142), plt.imshow(magnitude_spectrum, cmap='gray'), plt.title('Magnitude Spectrum')
plt.subplot(143), plt.imshow(np.log(1+np.abs(filtered_spectrum)), cmap='gray'), plt.title('Filtered Spectrum')
plt.subplot(144), plt.imshow(high_passed_image, cmap='gray'), plt.title('High Pass Filtered Image')
plt.show()
```
